```python
#video display
import numpy as np
import cv2

cap = cv2.VideoCapture(0)

#for saving the video
fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter('output.avi',fourcc, 20.0, (640,480))
 
while(True):
    ret, frame = cap.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    #recording each frame
    out.write(frame)
    
    cv2.imshow("frame",frame)
    #in imshow first parameter is name and the second is the frame 
    cv2.imshow('gray',gray)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
out.release()
cv2.destroyAllWindows()
```

```python
#logical operations
import numpy as np
import cv2


img1=cv2.imread("house.jpg",1)
img2=cv2.imread("python.png",1)

# I want to put logo on top-left corner, So I create a ROI
rows,cols,channels = img2.shape
roi = img1[0:rows, 0:cols ]

# Now create a mask of logo and create its inverse mask
img2gray = cv2.cvtColor(img2,cv2.COLOR_BGR2GRAY)

# add a threshold
ret, mask = cv2.threshold(img2gray, 220, 255, cv2.THRESH_BINARY_INV)

mask_inv = cv2.bitwise_not(mask)

# Now black-out the area of logo in ROI
img1_bg = cv2.bitwise_and(roi,roi,mask = mask_inv)

# Take only region of logo from logo image.
img2_fg = cv2.bitwise_or(img2,img2,mask = mask)

#adding im1bg to img2_fg
dst = cv2.add(img1_bg,img2_fg)
#replacing the first rows and columns with dst
img1[0:rows, 0:cols ] = dst

cv2.imshow("img2gay",img2gray)
cv2.imshow("mask",mask)
cv2.imshow("inv_mask",mask_inv)
cv2.imshow("img1_bg",img1_bg)
cv2.imshow("img2_fg",img2_fg)
cv2.imshow('res',img1)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Adaptive threshold is mainly used when the light on the image varies
#for best under standing of thresholding we need to use geeks for geeks only
import cv2
import numpy as np

img=cv2.imread("bookpage.jpg",1)
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

ret,th1=cv2.threshold(img,10,255,cv2.THRESH_BINARY)
retval, threshold2 = cv2.threshold(gray, 10, 255, cv2.THRESH_BINARY)
th3 = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 115, 1)
retval2,threshold4= cv2.threshold(gray,125,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)

cv2.imshow('img',img)
cv2.imshow("gray",gray)
cv2.imshow("threshold1",th1)
cv2.imshow('threshold2',threshold2)
cv2.imshow('Adaptive threshold',th3)
cv2.imshow('Otsu threshold',threshold4)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#filtering colors

import cv2

import numpy as np


cap=cv2.VideoCapture(0)

while True:
    ret,frame=cap.read()
    hsv=cv2.cvtColor(frame,cv2.COLOR_BGR2HSV)
    
    #defining the color range of pixels which we want to see
    min_value_of_pixel=np.array([0,0,0])
    max_value_of_pixel=np.array([150,150,150])
    
    #mask here  is a boolean.True when pixel is in range and False in other case
    mask=cv2.inRange(hsv,min_value_of_pixel,max_value_of_pixel)
    
    #finally displaying the frame with the mask earlier defined
    #There will be some pixel which we want tosee in the particular range of color and they will
    #have the value True in mask.Only those pixels  which are true in mask will be displaed 
    
    #mask use is to filter the frame or image only for particular pixels only like 
    #here displaying only the pixels of frame which have true value in max.
    result=cv2.bitwise_and(frame,frame,mask=mask)
    
    cv2.imshow("frame",frame)
    cv2.imshow("mask",mask)
    cv2.imshow("%result",result)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cv2.destroyAllWindows()
cap.release()
```

```python
#smoothing and blurring

import cv2
import numpy as np

cap=cv2.VideoCapture(0)

while True:
    ret,frame=cap.read()
    hsv=cv2.cvtColor(frame,cv2.COLOR_BGR2HSV)
    
    #defining the color range of pixels which we want to see
    min_value_of_pixel=np.array([0,0,0])
    max_value_of_pixel=np.array([150,150,150])
    
    #mask here  is a boolean.True when pixel is in range and False in other case
    mask=cv2.inRange(hsv,min_value_of_pixel,max_value_of_pixel)
    result=cv2.bitwise_and(frame,frame,mask=mask)
    
    #smoothing
    kernel = np.ones((15,15),np.float32)/225
    smoothed = cv2.filter2D(result,-1,kernel)
    #as the size of keranl increases decreases the noise around increases and clarity decreases
    #kernel1 = np.ones((5,5),np.float32)/25
    #smoothed1 = cv2.filter2D(result,-1,kernel1)
    
    #gaussian bluring
    #as the size of keranl increases decreases the noise around increases and clarity decrease
    blur = cv2.GaussianBlur(result,(15,15),0)
    
    #median blur
    #as the size of keranl increases decreases the noise around increases and clarity decrease
    median = cv2.medianBlur(result,5)
    
    #bilateral blur
    #as the size of keranl increases decreases the noise around increases and clarity decrease
    
    bilateral = cv2.bilateralFilter(result,15,75,75)
    
    cv2.imshow("frame",frame)
    #cv2.imshow("mask",mask)
#     cv2.imshow("result",result)
    cv2.imshow("smoothed",smoothed)
#     cv2.imshow("smoothed1",smoothed1)
    cv2.imshow('Gaussian Blurring',blur)
    cv2.imshow('Median Blur',median)
    cv2.imshow('bilateral Blur',bilateral)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cv2.destroyAllWindows()
cap.release()
```

```python
#morphological transformation
import cv2
import numpy as np

cap=cv2.VideoCapture(0)

while True:
    ret,frame=cap.read()
    hsv=cv2.cvtColor(frame,cv2.COLOR_BGR2HSV)
    
    #defining the color range of pixels which we want to see
    min_value_of_pixel=np.array([0,0,0])
    max_value_of_pixel=np.array([150,150,150])
    
    #mask here  is a boolean.True when pixel is in range and False in other case
    mask=cv2.inRange(hsv,min_value_of_pixel,max_value_of_pixel)
    result=cv2.bitwise_and(frame,frame,mask=mask)
    
    kernel = np.ones((5,5),np.uint8)
    erosion = cv2.erode(mask,kernel,iterations = 1)
    dilation = cv2.dilate(mask,kernel,iterations = 1)

    opening = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel)
    closing = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel)
    
    
    #tochat and blsckhat aare also the technicsof morphological trnsformation

#     cv2.imshow("mask",mask)
    cv2.imshow("frame",frame)
    cv2.imshow("result",result)
    cv2.imshow("erosion",erosion)
    cv2.imshow("dilation",dilation)
    cv2.imshow("opening",opening)
    cv2.imshow("closing",closing)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cv2.destroyAllWindows()
cap.release()
```
